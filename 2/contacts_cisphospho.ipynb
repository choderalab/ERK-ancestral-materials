```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
```

```python
contacts_wt = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_wt.npy')
contacts_mut = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_q102m.npy')
```

```python
contacts_wt.shape, contacts_mut.shape
```

```python
contacts_diff = contacts_mut - contacts_wt
```

```python
# residues of interest: V186 - L168, T188 - D147, T188 - K149, Y185 - D147, T183-D147
```

```python
np.sum(np.abs(contacts_diff[183]))
```

```python
changes =  []

for i in range(357):
    changes.append(np.sum(np.abs(contacts_diff[i])))
```

```python
changes_noabs =  []

for i in range(357):
    changes_noabs.append(np.sum(contacts_diff[i]))
```

```python
plt.figure(dpi=300)
plt.scatter(range(357),changes,s=10)
plt.plot(changes, linewidth=1)
plt.xlabel('Residue index')
plt.ylabel('Absolute sum of contact change')
plt.show()
```

```python
plt.figure(dpi=300)
plt.scatter(range(357),changes_noabs,s=10)
plt.plot(changes_noabs, linewidth=1)
plt.xlabel('Residue index')
plt.ylabel('Sum of contact change')
plt.show()
```
