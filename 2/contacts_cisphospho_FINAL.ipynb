```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
```

```python
erk2_wt = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_wt.npy')
erk2_55n = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_55n.npy')
erk2_q103m = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_q102m.npy')
erk2_55n_q103m = np.load('/Users/rafalpwiewiora/repos/ERK/contact_maps/contact_map_erk2_55n_q102m.npy')
```

```python
erk2_55n_wtsize = np.zeros((357,357))

for i in range(357):
    for j in range(357):
        if i >= 54:
            i_ = i+1
        else:
            i_ = i
        if j >= 54:
            j_ = j+1
        else:
            j_ = j
        erk2_55n_wtsize[i,j] = erk2_55n[i_,j_]
        
erk2_55n = erk2_55n_wtsize        
```

```python
erk2_55n_q103m_wtsize = np.zeros((357,357))

for i in range(357):
    for j in range(357):
        if i >= 54:
            i_ = i+1
        else:
            i_ = i
        if j >= 54:
            j_ = j+1
        else:
            j_ = j
        erk2_55n_q103m_wtsize[i,j] = erk2_55n_q103m[i_,j_]
        
        
erk2_55n_q103m = erk2_55n_q103m_wtsize        
```

```python
# 55N
```

```python
contacts_diff = erk2_55n - erk2_wt

changes =  []

for i in range(357):
    changes.append(np.sum(np.abs(contacts_diff[i])))
    
changes_noabs =  []

for i in range(357):
    changes_noabs.append(np.sum(contacts_diff[i]))
```

```python
plt.figure(dpi=300)
plt.scatter(range(357),changes,s=10)
plt.plot(changes, linewidth=1)
plt.xlabel('Residue index')
plt.ylabel('Absolute sum of contact change')
plt.show()
```

```python
plt.figure(dpi=300)
plt.scatter(range(357),changes_noabs,s=10)
plt.plot(changes_noabs, linewidth=1)
plt.xlabel('Residue index')
plt.ylabel('Sum of contact change')
plt.show()
```
